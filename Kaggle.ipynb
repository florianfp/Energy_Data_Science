```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
import time
from sklearn import tree
from sklearn import ensemble
from sklearn import cross_validation
```

```python
#I would like to parse the NaN directly as NaN and convert it into a 0/1
df = pd.read_csv('Training_set.csv', sep=';', header=0, names=['Index', 'Type', 'Date', 'Start_time', 'End_time', 'Usage', 'Units', 'Note', 
                 'Weekday', 'Temperature', 'Humidity', 'Visibility'], dtype={'Note':str})
df = df.drop(['Type', 'Units'], axis=1);
```

```python
#create groups to get day statistics 
df_groupby = df.groupby(by=['Weekday','Start_time'])

#quartiles
upper_quant = df_groupby.quantile(q=0.75)
lower_quant = df_groupby.quantile(q=0.25)
median_quant = df_groupby.quantile(q=0.5)

#weekdays and weekend mean
df_mean = df_groupby.mean()
df_mean_weekday = df_mean.loc[1:5,:]
df_mean_weekday = df_mean_weekday.mean(level=['Start_time'])
df_mean_weekend = (df_mean.loc[0,:] + df_mean.loc[6,:])/2
```

```python
#plot to show daily patterns and weekdays vs weekend
color = dict({0:'#300000', 1:'#200000', 2:'#FF0000', 3:'#780000', 4:'#300000', 5:'#300000 ', 6:'#FF0000'})
day = dict({0:'Sunday', 1:'Monday', 2:'Tuesday', 3:'Wednesday', 4:'Thursday', 5:'Friday', 6:'Saturday'})

fig, axes = plt.subplots(nrows=1, ncols=2)
#plot the quartile and median of each weekday (actually only three of them for clarity purposes)
plt.subplot(2, 1, 1)
for i in range(2,5):
    plt.plot(lower_quant[lower_quant['Weekday']==i]['Usage'], linewidth=0.2, color=color[i])
    plt.plot(upper_quant[lower_quant['Weekday']==i]['Usage'], linewidth=0.2, color=color[i])
    plt.plot(median_quant[median_quant['Weekday']==i]['Usage'], linewidth=1, color=color[i], label=day[i]+' median (with quartiles)')
#plot the mean for the week
plt.plot(df_mean_weekday['Usage'], label='Week days average', linestyle='dashed')
plt.title('Energy trend on weekdays')
plt.xlabel('Hour')
plt.ylabel('Consumption (kWh)')
plt.grid()
plt.legend(fontsize='medium', bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim([0,24]);

#plot the same values for the weekend
plt.subplot(2, 1, 2)
for i in [6,0]:
    plt.plot(lower_quant[lower_quant['Weekday']==i]['Usage'], linewidth=0.2, color=color[i])
    plt.plot(upper_quant[lower_quant['Weekday']==i]['Usage'], linewidth=0.2, color=color[i])
    plt.plot(median_quant[median_quant['Weekday']==i]['Usage'], linewidth=1, color=color[i], label=day[i]+' median (with quartiles)')
plt.legend()
#plot the mean for the week
plt.plot(df_mean_weekend['Usage'], label='Week end average', linestyle='dashed')
plt.title('Energy trend on weekend')
plt.xlabel('Hour')
plt.ylabel('Consumption (kWh)')
plt.grid()
plt.legend(fontsize='medium', bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlim([0,24]);

fig.tight_layout()
```

```python
#transform the date string feature from a string to three integer features
for i in range(0,len(df['Date'])):
        [day, month, year] = df.loc[i,'Date'].split('/')
        df.loc[i,'Day'] = day
        df.loc[i,'Month'] = month
        df.loc[i,'Year'] = year
        
#transform the time feature from a string to one integer feature
for i in range(0,len(df['Start_time'])):
    [hour,minutes] = df.loc[i,'Start_time'].split(':')
    df.loc[i,'Hour']=hour
    
#removing the features we will not be using anymore
df = df.drop(['Date','Start_time','Note', 'End_time'], axis=1)

#removing null values
null_index = df[df.isnull().any(axis=1)].index
df = df.drop(null_index, axis=0)
```

```python
#formating the data in the right way for the models
train = df.values
X = train[:,2:len(train)]
Y = train[:,1]
```

```python
#cross validation
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, Y, test_size=0.9, random_state=0)
```

```python
#Simple Classification Tree
clf = tree.DecisionTreeClassifier(min_samples_split=10)
clf = clf.fit(X_train, y_train)
```

```python
output = clf.predict(X_test)
```

```python
plt.plot(output[1:72])
```

```python
#feature importance
column_name = df.columns
feature_importance = clf.feature_importances_
for i in range(0,len(feature_importance)):
    print("Feature "+str(column_name[i+2])+" has importance "+str(feature_importance[i]))
```

```python
scores = cross_validation.cross_val_score(clf, X, Y, cv=2)
```

```python
plt.plot((output[1:72]-y_test[1:72])/y_test[1:72],'r')
plt.title('Decision tree relative error on first 72 hours')
```

```python
clf = ensemble.RandomForestClassifier(n_estimators=25, min_samples_leaf=5, min_samples_split=5)
clf = clf.fit(X, Y)
```

```python
scores = cross_validation.cross_val_score(clf, X, Y, cv=20)
```

```python
#directly using the in-built crossvalidation function
#N represents the number of parameters we are trying (N-1 actually)
N = 4

#lines represent parameter min_samples_leaf and columns parameter min_samples_split
accuracy = np.zeros((N-1,N-1))

#a trick not to handle complicated index for the loop on the number of trees
increment=10

for i in range(1,N):
    for j in range(1,N):
            clf = ensemble.RandomForestClassifier(n_estimators=25, min_samples_leaf=i, min_samples_split=j)
            #clf = tree.DecisionTreeClassifier(min_samples_leaf=i, min_samples_split=j)
            clf = clf.fit(X, Y)
            scores = cross_validation.cross_val_score(clf, X, Y, cv=20)
            accuracy[i-1,j-1] = scores.mean()

accuracy
```
