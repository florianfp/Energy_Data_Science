# Data wrangling

```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
import time
from sklearn import tree
from sklearn import ensemble
from sklearn import cross_validation
from sklearn import linear_model
from sklearn.cluster import MiniBatchKMeans, KMeans
```

```python
#I would like to parse the NaN directly as NaN and convert it into a 0/1
df = pd.read_csv('Training_set.csv', sep=';', header=0, names=['Index', 'Type', 'Date', 'Start_time', 'End_time', 'Usage', 'Units', 'Note', 
                 'Weekday', 'Temperature', 'Humidity', 'Visibility'], dtype={'Note':str})
df = df.drop(['Type', 'Units'], axis=1);
```

```python
#create groups to get day statistics 
df_groupby = df.groupby(by=['Weekday','Start_time'])

#quartiles
upper_quant = df_groupby.quantile(q=0.75)
lower_quant = df_groupby.quantile(q=0.25)
median_quant = df_groupby.quantile(q=0.5)

#weekdays and weekend mean
df_mean = df_groupby.mean()
df_mean_weekday = df_mean.loc[1:5,:]
df_mean_weekday = df_mean_weekday.mean(level=['Start_time'])
df_mean_weekend = (df_mean.loc[0,:] + df_mean.loc[6,:])/2
```

```python
#transform the date string feature from a string to three integer features
for i in range(0,len(df['Date'])):
        [day, month, year] = df.loc[i,'Date'].split('/')
        df.loc[i,'Day'] = int(day)
        df.loc[i,'Month'] = int(month)
        df.loc[i,'Year'] = int(year)
        
#transform the time feature from a string to one integer feature
for i in range(0,len(df['Start_time'])):
    [hour,minutes] = df.loc[i,'Start_time'].split(':')
    df.loc[i,'Hour']=int(hour)
    
#removing the features we will not be using anymore
df = df.drop(['Date','Start_time','Note', 'End_time'], axis=1)

#removing null values
null_index = df[df.isnull().any(axis=1)].index
df = df.drop(null_index, axis=0)
```

# Tree and Random Forest regression models

```python
#formating the data in the right way for the models
train = df.values
X = train[:,2:len(train)]
Y = train[:,1]
```

### Classification tree

#### Cross-validation for parameter selection

```python
#create the KFold iterator for crossvalidation
N = len(X)
cv_iterations = 20
kf = cross_validation.KFold(n=N, n_folds=cv_iterations)
step_size = 20
error_mean = np.zeros((9,9))

for k in range(1,10,1):

    for j in range(1,10,1):
    
        #pre-allocate the table of errors 
        error = np.zeros(cv_iterations)
        i=0
    
        for train_index, test_index in kf:

            #split the data
            X_train, X_test = X[train_index], X[test_index]
            Y_train, Y_test = Y[train_index], Y[test_index]
    
            #fit the model to new training data
            clf = tree.DecisionTreeClassifier(min_samples_split=j*step_size, min_samples_leaf=k)
            clf = clf.fit(X_train, Y_train)
    
            #make predictions
            output = clf.predict(X_test)
    
            #calculate RMSE
            error[i] = np.sqrt((1/float(len(output))*(np.sum((output-Y_test)**2))))

            i += 1
    
        #print('Mean: ',error.mean(), ' Max: ', error.max())
        error_mean[j-1,k-1] = error.mean()
        j += 1
```

```python
min_error = error_mean[1].min()
jk = np.where(error_mean==min_error)
print('Optimal min_sample_split is: '+ str(jk[0]+1) + ' and optimal min_samples_leaf is: ' + str(jk[1]+1)+' for a RMSE of '+str(min_error))
min_samples_split_optimal = jk[0]+1
min_samples_leaf_optimal = jk[1]+1
```

## Ridge regression

```python
#create the KFold iterator for crossvalidation
N = len(X)
cv_iterations = 10
kf = cross_validation.KFold(n=N, n_folds=cv_iterations)

mean_error = np.zeros((2,100))
j=0

for alpha_value in range(1,100000,1000):
    
    #pre-allocate the table of errors 
    error = np.zeros(cv_iterations)
    i=0
    for train_index, test_index in kf:

        #split the data
        X_train, X_test = X[train_index], X[test_index]
        Y_train, Y_test = Y[train_index], Y[test_index]
    
        #fit the model to new training data
        clf = linear_model.Ridge(alpha=alpha_value, fit_intercept=True, normalize = False)
        clf.fit(X_train, Y_train)
    
        #make predictions
        output = clf.predict(X_test)
    
        #calculate RMSE
        error[i] = np.sqrt((1/float(len(output))*(np.sum((output-Y_test)**2))))
        i += 1
        
        #results for this given alpha
        #print('Mean: ',error.mean(), ' Max: ', error.max())
        
        #store the result
        mean_error[0,j] = j
        mean_error[1,j] = error.mean()
    
    j +=1
```

```python
min_error = mean_error[1].min()
best_j = mean_error[0,mean_error[1]==min_error]
best_alpha = best_j *1000
best_alpha = best_alpha[0]
best_alpha, min_error
```

```python
#create the KFold iterator for crossvalidation
N = len(X)
cv_iterations = 3
kf = cross_validation.KFold(n=N, n_folds=cv_iterations)
nb_clusters = 5

j=0
a=1
b=2000
step_size = 200

#pre-allocate the table of errors 
error = np.zeros(cv_iterations)
error_mean = np.zeros((((b-a)/step_size)+1,2))


for nb_clusters in range(a,b,step_size):
    i=0
    
    for train_index, test_index in kf:

        #split the data
        X_train, X_test = X3[train_index], X3[test_index]
        Y_test = X_test[:,0]
    
        #fit the model to new training data
        k_means = KMeans(init='k-means++', n_clusters=nb_clusters, n_init=10)
        k_means.fit(X3);
    
        k_means_clusters = k_means.cluster_centers_
    
        #make predictions. The returned result is the cluster the point belongs to
        output_cluster = k_means.predict(X_test)
        prediction = k_means_clusters[output_cluster]
        output = prediction[:,0]
    
    
        #calculate RMSE
        error[i] = np.sqrt((1/float(len(output))*(np.sum(((output)-Y_test)**2))))
        i += 1
    
    error_mean[j,0] = nb_clusters
    error_mean[j,1] = error.mean()
    j += 1
```
